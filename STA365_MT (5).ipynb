## Part 1: Metropolis-Hastings

**Metropolis-Hastings** targetting **stationary distribution** $p$ based on **proposal distribution** $q$ given below

$$q(\tilde x^{(t)} | x^{(t-1)}) = \left(\frac{2}{3}\right)^{1-{\tilde x^{(t)}}}\left(\frac{1}{3}\right)^{\tilde x^{(t)}} \quad
\textrm{ and } \quad p(x^{(t)}) = \left(\frac{1}{3}\right)^{1-{x^{(t)}}}\left(\frac{2}{3}\right)^{x^{(t)}}$$

has **transition kernel** $K$ of transition probabilities 

$K = \left[\begin{array}{cc}\Pr(x^{(t)}=0 \rightarrow x^{(t+1)}=0) & \Pr(x^{(t)}=1 \rightarrow x^{(t+1)}=0)\\\Pr(x^{(t)}=0 \rightarrow x^{(t+1)}=1)& \Pr(x^{(t)}=1 \rightarrow x^{(t+1)}=1)\end{array}\right] = \left[\begin{array}{cl} 2/3 & 1/6 \\ 1/3 & 5/6 \end{array}\right]$

## Part 1B (5%)

### What is the Metropolis-Hastings acceptance probability $\alpha_{mh}$ here?

#### *Double Click and Answer below in this cell*

<!-- You don't have to write this in LaTeX but it's probably easy to do so -->
Let $\tilde{x}^{(t)} = \tilde{x}$ and $x^{t-1} = x$ Then ...

$\alpha_{mh} = \min\Biggl\{1,\, \frac{p(\tilde{x})\,q(x \mid \tilde{x})}{p(x)\,q(\tilde{x} \mid x)}\Biggr\}.$

Since our proposals are not symmetric, plugging in the functional expansion above yeilds:

$\alpha_{mh} = \min\Biggl\{1,\, \frac{\frac{1}{3}^{1-\tilde{x}} \frac{2}{3}^{\tilde{x}} \times \frac{2}{3}^{1-x} \frac{1}{3}^{x}}{\frac{1}{3}^{1-x} \frac{2}{3}^{x} \times \frac{2}{3}^{1-\tilde{x}} \frac{1}{3}^{\tilde{x}}}\Biggr\}.$

which simplifies to: 

$\alpha_{mh} = \min\Biggl\{1,\, \frac{\frac{1}{3}^{1-2\tilde{x}} \frac{2}{3}^{2\tilde{x}-1}}{\frac{1}{3}^{1-2x} \frac{2}{3}^{2x-1}}\Biggr\}.$



### What is the difference between the *acceptance probability,  proposal distribution, transition kernel, stationary distribution and target distribution*?

#### *Double Click and Answer below in this cell*

The stationary distribution is a distribution emergent in convergent markov chains. Specifically it is the distribution of the states in the limit as you appply the transition kernel over and over. The transition kernel indicates how the probabilities of the states change from one point in time to the next in our markov process. The target distribution is what we want to sample from and is specified initially. If reversibility and detailed balance are established, then we have a convergent markov chain where the stationary distribution is the target distribution. The proposal distribution is a distribution which provides us proposal samples from which we reject or accept according to the acceptance probability. It covers the support of the target distribution as well. When the proposals are dependent they are centered (for normal proposals) at the previous value hence it satisfies the markov property. Finally the acceptance probability (embeeds the reversibility condition ensuring covergence) is the probability we accept a given proposal that we sample from the proposal distribution.

```python
# If u is a vector (with `shape=272` in this case)
# then `(1-u)*theta_w0 + u*theta_w1` creates a vector of length 272
# which can be used for the mean of the observed `w` data along with `tau=tau_w`
# so long as `shape=1` is used in the specification of `theta_w0` and `theta_w1`
# which makes these random variables a "vector" as opposed to a "scalar"


with pymc.Model() as OldFaithful:
    p = pymc.Beta("p", alpha=1, beta=1)
    u = pymc.Bernoulli("u", p=p, shape=272)
    
    theta_w0 = pymc.Normal("theta_w0", mu=55 , tau = (1/5**2))
    theta_w1 = pymc.Normal("theta_w1", mu=85 , tau = (1/5**2))
    tau_w = pymc.Gamma("tau_w", alpha = 1/2, beta = 100/2)
    
    wi = pymc.Normal("wi", mu= (1-u)*(theta_w0) + (u)*theta_w1, sigma=tau_w)
    
    
    theta_y0 = pymc.Normal("theta_y0", mu=2.0 , tau = (1/0.5**2))
    theta_y1 = pymc.Normal("theta_y1", mu=4.5 , tau = (1/0.5**2))
    tau_y = pymc.Gamma("tau_y", alpha = 1/2, beta = 1/2)
    
    yi = pymc.Normal("yi", mu= (1-u)*(theta_y0) + (u)*theta_y1, sigma=tau_y)
    
    dat = old_faithful
    # The observed data flows into the model through two entery points
    obs_w = pymc.Normal("obs_w", mu = (1-u)*theta_w0 + u*theta_w1, 
                        tau=tau_w, observed=dat.waiting)
    obs_y = pymc.Normal("obs_y", mu = (1-u)*theta_y0 + u*theta_y1, 
                        tau=tau_y, observed=dat.eruptions)
    
```

## Part 4D (10%): Complete Gibbs Sampler


```python
n = 272
G = 100

# Initialize parameters
thetas_w_mu0 = np.array([55, 85])
thetas_w = np.zeros((2, G)) + thetas_w_mu0.reshape(2, 1)
thetas_w_tau0 = 1 / 5**2

thetas_y_mu0 = np.array([2.0, 4.5])
thetas_y = np.zeros((2, G)) + thetas_y_mu0.reshape(2, 1)
thetas_y_tau0 = 1 / 0.5**2

taus_alpha = 1
taus_beta = np.array([10**2, 1**2])
taus = np.zeros((2, G)) + taus_alpha / taus_beta.reshape(2, 1)

us = np.zeros((n, G))
ps = np.zeros(G) + 0.5

# Gibbs sampling loop
for g in range(1, G):
    # --- Sample u_i | thetas_w, thetas_y, taus, ps ---
    for i in range(n):
        # Likelihoods for u_i=1 and u_i=0
        likelihood_u1 = (
            stats.norm.pdf(old_faithful.waiting[i], loc=thetas_w[1, g-1], scale=np.sqrt(1.0 / taus[0, g-1])) *
            stats.norm.pdf(old_faithful.eruptions[i], loc=thetas_y[1, g-1], scale=np.sqrt(1.0 / taus[1, g-1])) *
            ps[g-1]
        )
        likelihood_u0 = (
            stats.norm.pdf(old_faithful.waiting[i], loc=thetas_w[0, g-1], scale=np.sqrt(1.0 / taus[0, g-1])) *
            stats.norm.pdf(old_faithful.eruptions[i], loc=thetas_y[0, g-1], scale=np.sqrt(1.0 / taus[1, g-1])) *
            (1 - ps[g-1])
        )
        
        # Normalize probabilities
        prob_u1_normalized = likelihood_u1 / (likelihood_u1 + likelihood_u0)
        us[i, g] = stats.bernoulli.rvs(prob_u1_normalized)

    # --- Sample p | u ---
    alpha_post = 1 + (us[:, g] == 1).sum()
    beta_post = 1 + n - (us[:, g] == 1).sum()
    ps[g] = stats.beta.rvs(a=alpha_post, b=beta_post)

    # --- Sample theta_w0, theta_w1 ---
    mask0 = (us[:, g] == 0)
    mask1 = (us[:, g] == 1)
    
    # For theta_w0
    n0 = mask0.sum()
    sum_w0 = np.sum(old_faithful.waiting[mask0])
    post_prec_w0 = thetas_w_tau0 + n0 * taus[0, g-1]
    post_mean_w0 = (thetas_w_tau0 * thetas_w_mu0[0] + taus[0, g-1] * sum_w0) / post_prec_w0
    thetas_w[0, g] = stats.norm.rvs(loc=post_mean_w0, scale=np.sqrt(1.0 / post_prec_w0))

    # For theta_w1
    n1 = mask1.sum()
    sum_w1 = np.sum(old_faithful.waiting[mask1])
    post_prec_w1 = thetas_w_tau0 + n1 * taus[0, g-1]
    post_mean_w1 = (thetas_w_tau0 * thetas_w_mu0[1] + taus[0, g-1] * sum_w1) / post_prec_w1
    thetas_w[1, g] = stats.norm.rvs(loc=post_mean_w1, scale=np.sqrt(1.0 / post_prec_w1))

    # --- Sample theta_y0, theta_y1 ---
    sum_y0 = np.sum(old_faithful.eruptions[mask0])
    post_prec_y0 = thetas_y_tau0 + n0 * taus[1, g-1]
    post_mean_y0 = (thetas_y_tau0 * thetas_y_mu0[0] + taus[1, g-1] * sum_y0) / post_prec_y0
    thetas_y[0, g] = stats.norm.rvs(loc=post_mean_y0, scale=np.sqrt(1.0 / post_prec_y0))

    sum_y1 = np.sum(old_faithful.eruptions[mask1])
    post_prec_y1 = thetas_y_tau0 + n1 * taus[1, g-1]
    post_mean_y1 = (thetas_y_tau0 * thetas_y_mu0[1] + taus[1, g-1] * sum_y1) / post_prec_y1
    thetas_y[1, g] = stats.norm.rvs(loc=post_mean_y1, scale=np.sqrt(1.0 / post_prec_y1))

    # --- Sample tau_w, tau_y ---
    # For tau_w (waiting times)
    sum_sq_w = (
        np.sum((old_faithful.waiting[mask0] - thetas_w[0, g])**2) +
        np.sum((old_faithful.waiting[mask1] - thetas_w[1, g])**2)
    )
    shape_post_w = taus_alpha + n / 2
    rate_post_w = taus_beta[0] + 0.5 * sum_sq_w
    taus[0, g] = stats.gamma.rvs(a=shape_post_w, scale=1.0 / rate_post_w)

    # For tau_y (eruptions)
    sum_sq_y = (
        np.sum((old_faithful.eruptions[mask0] - thetas_y[0, g])**2) +
        np.sum((old_faithful.eruptions[mask1] - thetas_y[1, g])**2)
    )
    shape_post_y = taus_alpha + n / 2
    rate_post_y = taus_beta[1] + 0.5 * sum_sq_y
    taus[1, g] = stats.gamma.rvs(a=shape_post_y, scale=1.0 / rate_post_y)
```
